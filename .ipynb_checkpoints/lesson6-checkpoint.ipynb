# Histogram and Scatterplots

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def symbol_to_path(symbol, base_dir = "Data"):
    """Return CSV path with given ticker symbol"""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))
def get_data(symbols, dates):
    df = pd.DataFrame(index=dates)
    """Read stock data (adjusted close) for given symbols from CSV files."""
    if 'SPY' not in symbols:
        """add SPY for reference"""
        symbols.insert(0, 'SPY') 
        
    for symbol in symbols:
        df_temp = pd.read_csv(symbol_to_path(symbol), index_col = "Date", 
                          parse_dates = True, usecols = ['Date', 'Adj Close'], na_values = ['nan'])
        df_temp.rename(columns={'Adj Close': symbol}, inplace=True)
        """column name must be unique"""
        df = df.join(df_temp)
        df = df.dropna()
    
    return df

def plot_data(df, title="Stock prices"):
    """Plot stock prices with a custom title and meaningful axis labels."""
    ax = df.plot(title=title, fontsize=12)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()

def cal_daily_returns(df):
    daily_returns = (df/df.shift(1))-1
    daily_returns.iloc[0,:]=0
    return daily_returns
    
def test_run():
    # Define a date range
    sd,ed = '2020-01-30', '2020-07-30'
    dates = pd.date_range('2020-01-30', '2020-07-30')

    # Choose stock symbols to read
    symbols = ['AAPL', 'MFT', 'TSLA', 'GOOG']
    
    # Get stock data
    df = get_data(symbols, dates)
    
    # Slice and plot
    ax = df['SPY'].plot(title = "SPY rolling mean", label = 'SPY')
    
    rm_SPY = df['SPY'].rolling(window=20).mean()
    rm_SPY.plot(label="Rolling Mean",ax=ax)
    
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    ax.legend(loc = 'upper left')
    plt.show()
    #define the window is 20 days so the first 20 days value are empty
    
def plot_selected(df, columns, start_index, end_index):
    """Plot the desired columns over index values in the given range."""
    df = normalize_data(df)
    plot_data(df.loc[start_index:end_index, columns])

def normalize_data(df):
    """Normalize stock prices using the first row of the dataframe."""
    return df / df.iloc[0, :]
```

```python
sd,ed = '2020-01-30', '2020-07-30'
dates = pd.date_range('2020-01-30', '2020-07-30')

# Choose stock symbols to read
symbols = []

# Get stock data
df = get_data(symbols, dates)
df.head()
plot_data(df)
daily_returns = cal_daily_returns(df)
daily_returns
```

```python
plot_data(daily_returns, title="Daily Returns")
```

```python
daily_returns.hist()
#default value of bins is 10
```

# Compare - Histogram
We calculate the histogram from the DAILY RETURNS of the two stocks. It gives us a sense how the daily returns compare to each other

```python
dates = pd.date_range(sd, ed)
df = get_data(symbols=['TSLA'],dates=dates)
df.head()
```

```python
daily_returns=cal_daily_returns(df)
plot_data(daily_returns,title="Daily Returns")
```

```python
daily_returns.hist(bins=20)
plt.show()
```

```python
daily_returns['SPY'].hist(bins=20,label='SPY')
daily_returns['TSLA'].hist(bins=20,label='TSLA')
plt.legend(loc='upper right')
plt.show()
```

A lower peak corresponds to higher volitility as the stocks are distributed over a higher range of prices. A right shift of the gaussian corresponds to a higher return, since the mean is higer.


Kurtosis is the thickness of the tails of the histogram. It is used to compare the stock to an ideal gaussian.

How much different of the distribution from the regular normal distribution 
fat tails: postitive kurtosis

# Compare - Scatterplot

```python
dates = pd.date_range(sd,ed)
df = get_data(symbols=['TSLA','MFT'],dates=dates)
df = normalize_data(df)
df.plot()
plt.show()
```

```python
daily_returns=cal_daily_returns(df)
plot_data(daily_returns,title="Daily Returns")
```

```python
daily_returns.plot(kind='scatter',x='SPY',y='TSLA')
plt.show()
```

```python
daily_returns.plot(kind='scatter',x='SPY',y='MFT')
plt.show()
```

```python
beta_TSLA , alpha_TSLA = np.polyfit(daily_returns['SPY'],daily_returns['TSLA'],1)
print(beta_TSLA , alpha_TSLA)
```

```python
daily_returns.plot(kind='scatter',x='SPY',y='TSLA')
plt.plot(daily_returns['SPY'],daily_returns['SPY']*beta_TSLA + alpha_TSLA,'-',color='r')
plt.show()
```

```python
beta_MFT , alpha_MFT = np.polyfit(daily_returns['SPY'],daily_returns['MFT'],1)
print(beta_MFT , alpha_MFT)
```

```python
daily_returns.plot(kind='scatter',x='SPY',y='MFT')
plt.plot(daily_returns['SPY'],daily_returns['SPY']*beta_MFT+ alpha_MFT,'-',color='r')
plt.show()
```
